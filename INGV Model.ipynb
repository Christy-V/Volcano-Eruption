## Normalizing the Dataset

Since the magnitude of values range across the dataset, We are Normalizing the  data set using min max scaler

````
# Since the range of values across different columns are different we are Normalizing the Dataset 
# https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html
# https://www.geeksforgeeks.org/standardscaler-minmaxscaler-and-robustscaler-techniques-ml/
scaler=MinMaxScaler()
scaled = scaler.fit(X_train)
filename = 'minmax.pkl'
pickle.dump(scaled, open(filename, 'wb'))
loaded_model = pickle.load(open('minmax.pkl', 'rb'))
scaled_tr = loaded_model.transform(X_train)
X_train = pd.DataFrame(scaled_tr, columns =X_train.columns.values )
scaled_cv=loaded_model.transform(X_cv)
X_cv = pd.DataFrame(scaled_cv, columns =X_cv.columns.values )
scaled_test=loaded_model.transform(X_test)
X_test = pd.DataFrame(scaled_test, columns =X_test.columns.values )
````

## LGBM  (Ensemble Model)

Since the metric high require lower errors and high errors could cost lives. I gave a shot on ensemble model too.

### With All features

````
#initialising the model
#https://www.kaggle.com/amanooo/ingv-volcanic-basic-solution-stft
gbdt =lgb.LGBMRegressor(boosting_type ='gbdt', objective ='rmse',silent = True)
````

````
# list of experimenting parameters
# https://lightgbm.readthedocs.io/en/latest/pythonapi/lightgbm.LGBMRegressor.html
Params = {'num_leaves': [10,30,20],'learning_rate': [0.01, 0.05, 0.1],'max_depth': [5,10,20],'n_estimators': [50, 100, 200],'min_data_in_leaf': [5, 10, 20]}
````

````
#https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html
grsrch = GridSearchCV(gbdt, Params,cv=3, verbose=1, n_jobs=-1)
````

````
# Searching for the best parameters
grsrch.fit(X_train, y_train)
````

````
# best parameters
grsrch.best_params_
````

````
# initialising the model with best parameters
gbdt_best =lgb.LGBMRegressor(boosting_type ='gbdt', objective ='rmse',learning_rate= 0.01, max_depth=20,min_data_in_leaf = 10,n_estimators = 200, num_leaves=30, silent = True)
````

````
# training the model
gbdt_best.fit(X_train,y_train)
````

````
filename = 'lgbm.pkl'
pickle.dump(gbdt_best, open(filename, 'wb'))
````

````
loaded_model = pickle.load(open('lgbm.pkl', 'rb'))
````

````
# predicting the values for train data
y_pred = loaded_model.predict(X_train)
````

````
# calculating the Root mean square
rmse=mean_absolute_error(y_train, y_pred)
````

````
# train RMSE
rmse
````

````
# predicting the values for validation data
y_pred = gbdt_best.predict(X_cv)
````

````
# calculating the Root mean square validation data
rmse=mean_absolute_error(y_cv, y_pred)
````

````
#Validation RMSE
rmse 
````

````
# predicting the values for test data
y_pred = gbdt_best.predict(X_test)
````

````
# calculating the Root mean square validation data
rmse=mean_absolute_error(y_test, y_pred)
````

````
#Test RMSE
rmse
````
